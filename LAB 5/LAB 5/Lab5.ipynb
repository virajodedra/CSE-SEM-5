```python
# actual start to remove the columns having the misisng values
#  here the data loss is very high so we can not use this 
# data.dropna()


# df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
#  here how = "any " if there any null valaue found it removes whole column but "all" is for like there is all null values in the column
dup_data = pd.read_csv("titanic.csv")
dup_data = dup_data.dropna(how = 'all', axis = 0)
dup_data
```

```python
#  here if we directly use fillna('ABC') it fill the blank column with the ABC it can change the data type of the column so the solution is belowed shown

data_copy = data.copy()
data_fillna = data_copy.fillna({
    'Age' : 35,
    'Cabin' : "Not Available"
})
data_fillna
data_fillna.info()
```

```python
meanAge = data.Age.mean()
meanAge
# here mode()[0] returns most frequent value
modeCabin = data.Cabin.mode()[0] 
#  here we get 2 values in the mode()[0] so to get the first we apply split()[0] here
modeCabin = modeCabin.split()[0] 

data_fillna = data.fillna({
    'Age' : meanAge.round(0),
    'Cabin': modeCabin
})
data_fillna
```

```python
# interpolate only works in the numeric column
# interpolate() is used to estimate missing values in a numeric column based on existing data. 
# data.interpolate(method='linear', axis=0, inplace=False, limit=None, limit_direction='forward')


data_with_interpolate = pd.read_csv('titanic.csv')
data_with_interpolate = data.interpolate()
data_with_interpolate
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
#  scaling refer to range, main used  for accuracy  
# used to find outliers

pureData = data.copy()
pureData['Age'] = pureData['Age'].interpolate()
# pureData
age_pure = pureData['Age']
age_min =  pureData['Age'].min()
age_max = pureData['Age'].max()
pureData['MinMax_Age'] = ( age_pure -  age_min) / ( age_max - age_min )
pureData[['Age', 'MinMax_Age']]
```

```python

pureData2 = data.copy()
pureData2['Age'] = pureData2['Age'].interpolate()
# Age = pureData2.Age
# pureData2 = pureData2.fillna({
#     'Age' : Age.round(0)
# })

max_age_count = len(str(int(pureData2.Age.max())))
j = max_age_count
print(max_age_count)

pureData2['Age_scaled'] = pureData2['Age'] / (10 ** j)
pureData2[['Age', 'Age_scaled']]
```

```python
mean_age = pureData['Age'].mean()
age_std = pureData['Age'].std()

pureData2['Age_zscore'] = (pureData2['Age'] - mean_age) / age_std
print(pureData2[['Age', 'Age_scaled', 'Age_zscore']])
```
